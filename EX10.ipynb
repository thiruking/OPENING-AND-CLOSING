```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='THIRU',org=(100,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img

def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()

img = load_img()
display_img(img)
```

```python
white_noise = np.random.randint(
    low = 0,
    high = 3,
    size = (600,600)
)
```

```python
white_noise
```

```python
white_noise = white_noise*255
```

```python
white_noise
```

```python
noise_img = white_noise+img
display_img(noise_img)
```

```python
kernel = np.ones((7,7),dtype=np.uint8)
opening = cv2.morphologyEx(
    noise_img,
    cv2.MORPH_OPEN,
    kernel
)
```

```python
## opening image
display_img(opening)
```
